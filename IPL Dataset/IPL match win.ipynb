```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Loading IPL matches dataset

ipl=pd.read_csv("IPL Matches 2008-2020.csv")
```

```python
ipl
```

```python
#loading first five results
ipl.head()
```

```python
#looking for rows and columns in ipl

ipl.shape
```

```python
#Getting the frequency of the most man of the match

ipl["player_of_match"].value_counts()
```

```python
#Getting top 5 player with the most man of the match

top_5= ipl["player_of_match"].value_counts()[0:5]
top_5
```

```python
list(top_5.keys())
```

```python
plt.bar(list(top_5.keys()),list(top_5))
```

```python
#Getting frequency oof result column 
ipl["result"].value_counts()
```

```python
#Finding out number of toss wins w.r.t. each team
ipl["toss_winner"].value_counts()
```

```python
#Extracting records where the team won batting first
bating_first= ipl[ipl["result"]!="wickets"]
```

```python
bating_first.head()
```

```python
#Finding out number of win w.r.t.each team after batting first
bating_first["winner"].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(bating_first["winner"].value_counts()[0:3].keys(),bating_first["winner"].value_counts()[0:3],color=["grey","pink","blue"])
```

```python
plt.figure(figsize=(6,6))
plt.pie(bating_first["winner"].value_counts(),labels=bating_first["winner"].value_counts().keys(),autopct="%0.1f%%")
```

```python
# Extracting those records where a team has won after bating second
bating_2= ipl[ipl["result"]!="runs"]
```

```python
bating_2.head()
```

```python
bating_2["winner"].value_counts()
```

```python
bating_2["winner"].value_counts()[0:5]
```

```python
plt.figure(figsize=(7,7))
plt.bar(bating_2["winner"].value_counts()[0:3].keys(),bating_2["winner"].value_counts()[0:3])
```

```python
plt.figure(figsize=(7,7))
plt.pie(bating_2["winner"].value_counts(),labels= bating_2["winner"].value_counts().keys(),autopct="%0.1f%%")
```
